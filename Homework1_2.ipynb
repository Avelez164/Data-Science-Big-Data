```python
import statsmodels.api as sm

mtcars_dataset = sm.datasets.get_rdataset("mtcars")
mtcars = mtcars_dataset.data
```

```python
num_rows = mtcars.shape[0]
num_rows
```

b. How many columns are there?

```python
num_columns = mtcars.shape[1]
num_columns
```

```python
first_ten = mtcars.head(10)
first_ten
```

f. Show every other  row (i.e., 1st, 3rd, 5th, …)

```python
every_other = mtcars.iloc[::2]
every_other
```

g. What is the mean mpg value

```python
mean_mpg = mtcars['mpg'].mean()
mean_mpg
```

h. Show all rows where the number of cylinders is 6

```python
six_cyl = mtcars[mtcars['cyl'] == 6]
six_cyl
```

i. Show all rows where its mpg is lower than the mean mpg value

```python
mpg_lower = mtcars[mtcars['mpg'] < mean_mpg]
mpg_lower
```

j. What is the horsepower of the car with the highest mpg (code should show only the horsepower value)?

```python
highest_mpg_hp = mtcars.loc[mtcars['mpg'].idxmax(), 'hp']
highest_mpg_hp
```

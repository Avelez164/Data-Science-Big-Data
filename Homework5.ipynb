1. Consider the billboard dataset that is supplied with the tidyr package which shows the Billboard top 100 song rankings in the year 2000. Apply the Pandas data wrangling methods to answer these questions. For each question, give only the code.

```python
import pandas as pd
import statsmodels.api as sm

billboard_dataset = sm.datasets.get_rdataset("billboard", "tidyr")
billboard = billboard_dataset.data

billboard.melt(
    id_vars=['artist', 'track', 'date.entered'],
    var_name='wk',
    value_name='rank'

).dropna()
```

a) Show for each track, how many weeks it spent on the chart

```python
(
    billboard
    .groupby('track')['date.entered']
    .count()
)
```

b) List tracks in decreasing order of number of weeks spent on the chart


```python
(
    billboard
    .groupby('track')['date.entered']
    .count()
    .sort_values(ascending=False)
)
```

c) Show for each track, its top rank

```python
(
    billboard
    .groupby('track')
    .agg(min_rank=('wk1', 'min'))
)
```

d) List tracks in increasing order of its top rank

```python
(
    billboard
    .groupby('track')
    .agg(min_rank=('wk1', 'min'))
    .sort_values(by='min_rank', ascending=True)
)
```

e) Show for each artist, their top rank

```python
(
    billboard
    .groupby('artist')
    .agg(min_rank=('wk1', 'min'))
)
```

f) List artists in increasing order of their top rank

```python
(
    billboard
    .groupby('artist')
    .agg(min_rank=('wk1', 'min'))
    .sort_values(by='min_rank', ascending=True)
)
```

g) List tracks that spent more than 35 weeks in the charts

```python
(
    billboard
    .groupby('track')
    .agg(total_weeks=('wk1', 'count'))
    .query('total_weeks > 35')
)
```

h) List tracks that spent more than 35 weeks in the charts along with their artists

```python
(
    billboard
    .groupby(['track', 'artist'])
    .agg(total_weeks=('wk1', 'count'))
    .query('total_weeks > 35')
)
```

**2**. The demographics.csv  file (available in the Tidy data module on Canvas) gives the proportion of a country’s population in different age groups and some other demographic data such as mortality rates and expected lifetime.

a) The data is not “tidy”. In 2-3 sentences, explain why.

**The data is not tidy because the demographic data is spread across multiple columns, for every column there's a specific variable label for it. The data also includes the Series name as a column, this has to be dropped. Male and Female are also stored in seperate columns we should be able to observe one per row.**

b) Transform the table to tidy data with one country per row. [Give code]

```python
import pandas as pd


url = "https://drive.google.com/uc?export=download&id=1RU6T6MOLvOKE0m0rezm5iRbXI4lRPgka"

df = pd.read_csv(url)

unique_series_names = df["Series Name"].unique()

# Display the unique Series Names
for name in unique_series_names:
    print(name)
```

```python
df_wide = df.pivot(
    index=["Country Name", "Country Code"],
    columns="Series Name",
    values="YR2015"
).reset_index()

df_wide = df_wide.dropna()

# Convert all numeric columns to numeric type (ignore errors)
for col in df_wide.columns[2:]:  # Skip 'Country Name' and 'Country Code'
    df_wide[col] = pd.to_numeric(df_wide[col], errors="coerce")

# Display the first few rows of the wide data
print(df_wide)
```

c) Add the male/female population numbers together (i.e., ignore sex-related differences). [Hint: You will have to mutate for every pair of columns, e.g., .assign(**{'SP.POP.0014.IN': lambda df: df['SP.POP.0014.MA.IN'] + df['SP.POP.0014.FE.IN']})] [Give code]

```python
# Print the column names
print(df_wide.columns)
```

```python
if all(col in df_wide.columns for col in ["SP.POP.0014_IN", "SP.POP.1564_IN", "SP.POP.65UP_IN", "SP.POP.80UP_IN", "SP.POP.TOTL_IN"]):
    print("The male/female population numbers have already been combined.")
else:
    print("The combined population columns are missing. Please check the dataset.")
```

```python
# Combine male/female population numbers
df_wide = df_wide.assign(
    Population_0_14=lambda df: pd.to_numeric(df["Population ages 0-14, male"], errors="coerce") + pd.to_numeric(df["Population ages 0-14, female"], errors="coerce"),
    Population_15_64=lambda df: pd.to_numeric(df["Population ages 15-64, male"], errors="coerce") + pd.to_numeric(df["Population ages 15-64, female"], errors="coerce"),
    Population_65_above=lambda df: pd.to_numeric(df["Population ages 65 and above, male"], errors="coerce") + pd.to_numeric(df["Population ages 65 and above, female"], errors="coerce"),
    Population_80_above=lambda df: pd.to_numeric(df["Population ages 80 and above, male"], errors="coerce") + pd.to_numeric(df["Population ages 80 and above, female"], errors="coerce"),
    Population_Total=lambda df: pd.to_numeric(df["Population, male"], errors="coerce") + pd.to_numeric(df["Population, female"], errors="coerce")
)

# Drop the original male/female columns
df_wide = df_wide.drop(columns=[
    "Population ages 0-14, male", "Population ages 0-14, female",
    "Population ages 15-64, male", "Population ages 15-64, female",
    "Population ages 65 and above, male", "Population ages 65 and above, female",
    "Population ages 80 and above, male", "Population ages 80 and above, female",
    "Population, male", "Population, female"
])

# Display the resulting DataFrame
print(df_wide.head())
```

d) Write code to show the top 5 countries with the lowest proportion of the population below 14 years old (i.e., SP.POP.0014.IN/SP.POP.TOTL) [Code, and list of 5 countries]

```python
# Ensure the columns are numeric
df_wide["Population_0_14"] = pd.to_numeric(df_wide["Population_0_14"], errors="coerce")
df_wide["Population_Total"] = pd.to_numeric(df_wide["Population_Total"], errors="coerce")

# Calculate the proportion of the population below 14 years old
df_wide["Prop_Below_14"] = df_wide["Population_0_14"] / df_wide["Population_Total"]

# Find the top 5 countries with the lowest proportion
top_5_lowest = df_wide[["Country Name", "Prop_Below_14"]].nsmallest(5, "Prop_Below_14")

# Display the result
print(top_5_lowest)
```

## Part 1 - Slide 18

From the iris dataset, show only Sepal.Length and Sepal.Width columns for those flowers with Petal.Length > 3.0 and Petal.Width > 1.0

```python
import statsmodels.api as sm

iris_dataset = sm.datasets.get_rdataset("iris")
iris = iris_dataset.data
```

```python
(
    iris
    .query('`Petal.Length` > 3.0 and `Petal.Width` > 1.0')
    .filter(['Sepal.Length', 'Sepal.Width'])
)
```

```python

(
    iris
    .groupby('Species')
    .agg({'Sepal.Length': 'max'})
)
```

```python
iris.filter(['Sepal.Length', 'Sepal.Width', 'Petal.length', 'Petal.Width'])
iris.drop('Species', axis=1)
```

```python
(
    iris
    .query('Species == "setosa"')
    .filter(['Sepal.Width'])
)
```

```python
#sorts in decending value
#lambda col: return -col
iris.sort_values(['Sepal.Length'], key=lambda col: -col)

```

## Part 2 - Slide 21

```python
mpg_dataset = sm.datasets.get_rdataset("mpg", "ggplot2")
mpg = mpg_dataset.data
```

```python
(
    mpg
    .groupby(['manufacturer', 'year'])
    .agg({'cty': 'mean'})
)
```

Show only manufacturer and cty columns in increasing order of cty

```python
mpg[['manufacturer', 'cty']].sort_values(by='cty')
#
(
    mpg
    .filter(['manufacturer', 'cty'])
    .sort_values(by='cty')
)
```


 Show only manufacturer, cty, hwy columns in increasing order of cty

```python
(
    mpg
    .filter(['manufacturer', 'cty', 'hwy'])
    .sort_values(by='cty')
)
```

Show only manufacturer, cty, hwy columns ordered first by cty, then by hwy

```python
(
    mpg
    .filter(['manufacturer', 'cty', 'hwy'])
    .sort_values(by=['cty', 'hwy'])
)
```

Show only manufacturer, cty, hwy columns in decreasing order of cty

```python
(
    mpg
    .filter(['manufacturer', 'cty', 'hwy'])
    .sort_values('cty', key=lambda city:-city)
)
```

Show only manufacturer, cty, hwy columns for cars made by audi in decreasing order of cty

```python
(
    mpg
    .query('manufacturer == "audi"')
    .filter(['manufacturer', 'cty', 'hwy'])
    .sort_values('cty', key=lambda city:-city)
)
```

Show only manufacturer, cty, hwy columns for cars made by honda in decreasing order of cty and cty > 25

```python
(
    mpg
    .query('manufacturer == "honda" & cty > 25')
    .filter(['manufacturer', 'cty', 'hwy'] )
    .sort_values('cty', key=lambda city:-city)
)
```

Show only models of cars made by honda in decreasing order of cty and cty > 25

```python
(
    mpg
    .query('manufacturer == "honda" & cty > 25')
    .sort_values('cty', key=lambda city:-city)
    .filter(['model'])
)
```

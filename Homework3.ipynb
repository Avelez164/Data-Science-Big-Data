## Part 1

Load the *nycflights13* dataset (will have to specify the **nycflights13** package) which contains flight arrival and departure data in a table called flights. Apply Pandas data wrangling methods to answer these questions. For each question, **give only the (one line) code**.

1. List data only for flights that departed on February 12, 2013.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
flights_dataset = sm.datasets.get_rdataset('flights', 'nycflights13')
flights = flights_dataset.data
```

2. List data only for flights that were delayed (both arrival and departure) by more than 2 hours.


```python
(
    flights
    .query('dep_delay > 120 & arr_delay > 120')
)
```

3. List data only for flights that were delayed (either arrival or departure) by more than 2 hours.


```python
(
    flights
    .query('dep_delay > 120 | arr_delay > 120')
)
```

4. List data only for flights that were operated by United, American, or Delta.





```python
(
    flights
    .query('carrier.isin([ "UA", "AA", "DL"])')
    # I just used the most relevant data
    .loc[:, ["flight", "carrier", "origin", "dest", "dep_time", "arr_time", "air_time", "distance"]]
    .dropna()

)
```

5. Sort data in order of fastest flights (air_time).


```python
(
    flights
    #.sort_values("air_time")
    .sort_values('air_time', key=lambda col: col)
    .dropna(subset=["air_time"])
    .loc[:, ["flight", "carrier", "origin", "dest", "dep_time", "arr_time", "air_time", "distance"]]

)
```

6. Sort data in order of longest duration flights (air_time).


```python
(
    flights
    .sort_values("air_time", key=lambda col: col)
    .dropna(subset=["air_time"])
    .loc[:, ["flight", "carrier", "origin", "dest", "dep_time", "arr_time", "air_time", "distance"]]
)
```

7. Show only the origin and destination of flights sorted by longest duration flights.


```python
(
    flights
    .sort_values("air_time", key=lambda col: col)
    .dropna(subset=["air_time"])
    .loc[:, ["origin", "dest", "air_time"]]

)
```

8. Add a new variable that indicates the total delay (both departure and arrival delay).


```python
(
    flights
    .assign(total_delay=(flights.dep_delay + flights.arr_delay))
)
```

9. Show only the origin and destination of flights sorted by descending order of total delay.


```python
(
    flights
    .assign(total_delay=(flights.dep_delay + flights.arr_delay))
    .sort_values('total_delay', key=lambda col: -col)
    .loc[:, ["origin", "dest", "total_delay"]]
    .dropna(subset=["total_delay"])
)
```

10. Show only the origin and destination of 10 most delayed flights

    *Hint*: there are multiple ways of solving this. Some additional methods that you will find useful are `.iloc`, `head()`, `nlargest()`, `rank(method='min')`.


```python
(
    flights
    .assign(total_delay=(flights.dep_delay + flights.arr_delay))
    .sort_values('total_delay', key=lambda x: -x)
    .head(10)[['origin', 'dest']]
)
```

11. Show the average total delay for every departure city.


```python
(
    flights
    .assign(total_delay=(flights.dep_delay + flights.arr_delay))
    .groupby(['origin', 'dest'])
    .total_delay
    .agg('mean')
)
```

12. Show the average total delay for every departure-arrival city pair.


```python
(
    flights
    .assign(total_delay=(flights.dep_delay + flights.arr_delay))
    .groupby(["origin", "dest"])
    .agg({"total_delay": "mean"})
    .reset_index()
)
```

## Part 2

Apply the Pandas data wrangling methods to answer these questions. For each question, give only the code (as one method chain with multiple “verbs” one after the other) beginning with survey. ….

a. Show only rows where Statistics skill is 10

```python
import pandas as pd
url = "https://drive.google.com/uc?export=download&id=15e8B1DHJhFT1IqSTbhYq8YLLIhEKcsD1"
survey = pd.read_csv(url)
```

```python
(
    survey
    .query('Statistics == 10')
    .filter(['Statistics'])
)
```

b. Show only rows where both Statistics and Math skills are 10

```python
(
    survey
    .query('Statistics == 10 and Math == 10')
    .filter(['Statistics', 'Math'])
)
```

c. Show only the Semester, Communication, and Visualization columns where both Statistics and Math skills are 10

```python
(
    survey
    .query('Statistics == 10 and Math == 10')
    .filter(['Semester', 'Communication', 'Visualization'])
)
```

d. Show only the Semester, Communication, and Visualization columns in decreasing order of Statistics and then Math skills

```python
(
    survey
    .query('Statistics == 10 and Math == 10')
    .sort_values(['Statistics', 'Math'], ascending=False)
    .filter(['Semester', 'Communication', 'Visualization'])
)
```

e. Show only rows where Statistics skill is missing

```python
(
    survey
    .query('Statistics == 10 and Math == 10')
    .filter(['Semester', 'Communication', 'Visualization'])
)
```

f. Show for every student, only their Year, Semester, Math skill, Statistics skill, and the maximum of their Math and Statistics skills (Hint: use axis=1)

```python
(
    survey
    .assign(max_skill=survey[['Math', 'Statistics']].max(axis=1))
    .filter(['Year', 'Semester', 'Math', 'Statistics', 'max_skill'])
)
```

g. Show the median value of Computer Science skills

```python
(
    survey
    .filter(['ComputerScience'])
    .median()
)
```

h. Show the median value of Computer Science skills in Fall 2024

```python
(
    survey
    .query('Semester == "Fall2024"')
    .filter(['ComputerScience'])
    .median()
)
```

i. Show the median values of (each of) Computer Science, Math, and Statistics skills in every year

```python
(
  survey
  .groupby(['Semester'])
  .agg(
      dict(
          ComputerScience='median',
          Math='median',
          Statistics='median'
      )
  )
)
```

j. Show for every year, the median values of (each of) Computer Science, Math, and Statistics skills of students who have taken CPSC483 (TakenCPSC483 is Yes)

```python
(
    survey
    .query('TakenCPSC483 == "Yes"')
    .groupby('Semester')
    .agg(
      dict(
          ComputerScience='median',
          Math='median',
          Statistics='median'
      )
  )
)
```

k. Show the median values of (each of) Computer Science, Math, and Statistics skills for groups of students who have taken and not taken CPSC483 (Hint: groupby)

```python
(
    survey
    .groupby('TakenCPSC483')
    .agg(
      dict(
          ComputerScience='median',
          Math='median',
          Statistics='median'
      )
    )
)
```

l. Show the number of students who took the survey in every semester and year

```python
(
    survey
    .groupby(['Semester'])
    .agg(student_count=('Semester', 'count'))
)
```

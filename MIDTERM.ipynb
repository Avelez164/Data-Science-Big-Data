Calculate the mean of values: 1, 2, 5

```python
num = [1, 2, 5]
mean = sum(num) / len(num)
print(mean)
```

Calculate the magnitude of vector (1, 2, 5)

Magnitude = $\sqrt{1^2 + 2^2 + 5^2}$

```python
mag = (num[0]**2 + num[1]**2 + num[2]**2)**0.5
print(mag)
```

Normalize the vector

*Divide each component by the magnitude (and save to a new variable)*

```python
norm_vector = [x / mag for x in num]
print (norm_vector)
```

Calculate the magnitude of the normalized vector

```python
mag_norm = (norm_vector[0]**2 + norm_vector[1]**2 + norm_vector[2]**2)**0.5
print(mag_norm)
```

```python
import numpy as np
import statistics
nums = np.array(sorted([36, 42, 100, 28, 17, 12, 9, 4, 1]))


q1, median, q3 = statistics.quantiles(nums)
iqr = q3 - q1
lower_fence = q1 - (1.5 * iqr)
upper_fence = q3 + (1.5 * iqr)
outliers = nums[(nums < lower_fence) | (nums > upper_fence)]
inside_fences = nums[(nums >= lower_fence) & (nums <= upper_fence)]
lower_whisker = inside_fences[0]
upper_whisker = inside_fences[-1]

print(nums)
lower_fence, lower_whisker, q1, median, q3 , upper_whisker, upper_fence, outliers
```

```python
import matplotlib.pyplot as plt
import pandas as pd
s = pd.Series(nums)

df = pd.DataFrame(dict(nums=nums))

s.plot.box()
plt.show()

sns.boxplot(df, y='nums')

(
    ggplot(df)
    + geom_boxplot(aes(y='nums'))
)
```

Transform the dataset such that:

* The License column is split into two:
  - `LicenseOrReceipt`: a categorical variable with two categories "License" and "Receipt"

```python
horse_racing = pd.read_csv(
    "https://raw.githubusercontent.com/ProfAvery/cpsc375-classwork/refs/heads/main/datasets/horse_racing.csv"
)

horse_racing[['LicenseOrReciept', 'Number']] = horse_racing.License.str.split('#', expand=True)
horse_racing.LicenseOrReciept = horse_racing.LicenseOrReciept.astype('category')
horse_racing.Number = horse_racing.Number.astype('int')
horse_racing.drop('License', axis=1, inplace=True)

horse_racing[['Month', 'Day', 'Year']] = horse_racing.Expires.str.split('/', expand=True)
horse_racing.Month = horse_racing.Month.astype('int')
horse_racing.Day = horse_racing.Day.astype('int')
horse_racing.Year = horse_racing.Year.astype('int')
horse_racing.drop('Expires', axis=1, inplace=True)

horse_racing.Name = horse_racing.Name.str.replace('0', 'O')

horse_racing
```

Consider the problem of estimating Height given Weight and Age
* What are some linear models for this?

Dependent variable / output (y): Height

Independent / explanatory / input ($ x_1, x_2 $): Weight, Age

# 5. **K-nearest neighbors (k-nn)**

KNN with the iris dataset.

Use $ \frac{2}{3} $ of the examples for training, then report accuracy on the test set.

```python
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

iris = sm.datasets.get_rdataset("iris").data
X = iris.iloc[:, 0:4]
y = iris.iloc[:, 4]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=100)

mod = KNeighborsClassifier(n_neighbors=1)
mod.fit(X_train, y_train)

mod.score(X_test, y_test)
```

Add `predicted` column to test set.

```python
pred = pd.concat(
    [X_test, y_test, pd.Series(mod.predict(X_test), name='Predicted', index=y_test.index)],
    axis=1,
    join="outer",
).sort_index()

pred
```

Count correct and incorrect predictions.

```python
pred[['Species', 'Predicted']].value_counts()
```

```python
from sklearn.neighbors import KNeighborsClassifier
mod = KNeighborsClassifier(n_neighbors = 1)
mod.fit(X_train, y_train)
mod.predict(X_test)
1 - mod.score(X_test, y_test)
```

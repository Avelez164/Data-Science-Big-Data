1. The R library ggplot2 also includes some datasets, including the mpg dataset. This dataset can be retrieved from Python with statsmodels:

```python
import statsmodels.api as sm

mpg_dataset = sm.datasets.get_rdataset("mpg", "ggplot2")
mpg = mpg_dataset.data
```

```python
%pip install lets-plot
```

```python
from lets_plot import *

LetsPlot.setup_html()
```

a.  Plot a scatterplot of variables displ and cty.

```python
(
    ggplot(mpg)
    + geom_point(aes(x="displ", y="cty"))
)
```

b. Redraw the previous scatterplot but also add all these:

*   more descriptive x and y-axis labels,
*   a title that should be the names of all group members, and
*   set cty limits to (0,40) and displ limits to (0,10)

```python
(
    ggplot(mpg)
    + geom_point(aes(x="displ", y="cty")) +
    labs(
       title='City mpg vs Engine Displacement'
    ) +
    scale_x_continuous(limits=[0,10]) +
    scale_y_continuous(limits=[0,40])
)
```

c. Plot a scatterplot of variables displ and cty. Show variable year also.

```python
(
    ggplot(mpg)
    + geom_point(aes(x="displ", y="cty", color='year')) +
    labs(
       title='City mpg vs Engine Displacement'
    ) +
    scale_x_continuous(limits=[0,10]) +
    scale_y_continuous(limits=[0,40])
)
```

d. Plot a scatterplot of variables displ and cty. Show variables year and class also.


```python
(
    ggplot(mpg)
    + geom_point(aes(x="displ", y="cty", color='year', shape='class')) +
    labs(
       title='City mpg vs Engine Displacement'
    ) +
    scale_x_continuous(limits=[0,10]) +
    scale_y_continuous(limits=[0,40])
)
```

3. Use the same data from the previous question (after removing NAs) to generate the following graphs. All plots should use Lets-Plot. Include both the Python code and the plot itself

a. Plot a histogram of variable Math

```python
(
    ggplot(survey_cleaned)
    + geom_histogram(aes(x="Math"))
)
```

b. The plot above likely has x-axis “breaks” not aligned with the bars. Provide your own breaks to match the bars. Hint: use the breaks argument in scale_x_continuous().

```python
(
    ggplot(survey_cleaned)
    + geom_histogram(aes(x="Math")) +
    scale_x_continuous(breaks=(1, 11))
)
```

c. Plot a boxplot of variable Math

```python
(
    ggplot(survey_cleaned)
    + geom_boxplot(aes(x="Math"))
)
```

d. Plot a scatterplot of variables Math and Statistics.**bold text**

```python
(
    ggplot(survey_cleaned)
    + geom_point(aes(x="Math", y="Statistics"))
)
```

e. Redraw the previous scatterplot but also:
Add more descriptive x and y-axis labels, add a title that should be the names of all group members, set both x-axis and y-axis limits to (1,10), and make sure x-axis and y-axis breaks are aligned with the points.


```python
(
    ggplot(survey_cleaned)
    + geom_point(aes(x="Math", y="Statistics"))
    + labs(
        x='Math Score',
        y='Statistics Score',
        title='Math vs Statistics Score'
    )
    + scale_x_continuous(limits=[1, 10], breaks=(1, 11))
    + scale_y_continuous(limits=[1, 10], breaks=(1, 11))
)
```

f. Plot a boxplot of variable ComputerScience vs. Semester (Note: you can plot either horizontally or vertically)

```python
(
    ggplot(survey_cleaned)
    + geom_boxplot(aes(x="Semester", y="ComputerScience"))
)
```

g. Visualize the two categorical variables TakenCPSC483 and PlanCPSC483.

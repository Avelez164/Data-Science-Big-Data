1. Add a new variable that indicates the total delay (both departure and arrival delay).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
flights_dataset = sm.datasets.get_rdataset('flights', 'nycflights13')
flights = flights_dataset.data

(
    flights
    .assign(total_delay = lambda x: x.arr_delay + x.dep_delay)
)
```

```python
(
    flights
    .assign(total_delay = lambda x: x.arr_delay + x.dep_delay)
    .sort_values(by='total_delay', ascending=False)
    .head(10)
    .reset_index(drop=True)
)
```

*c*. Write code (as a Pandas method chain) to merge the two tables above and create a single table with the total number of responses in every semester as below (the order of rows/columns is not important): [Give code only]

| Year | Semester | Responses | Enrolled |
|------|----------|-----------|----------|
| 2025 | Spring   |62         |73        |
| 2024 | Fall     |55         |51        |



```python
enrollment = pd.DataFrame(dict(Year=[2024, 2025], Semester=['Fall', 'Spring'], Enrollment=[51, 73]))

responses_aggregated = pd.DataFrame({
    'Year': [2024, 2025],
    'Semester': ['Fall', 'Spring'],
    'Responses': [55, 62]
})

new_merged = enrollment.merge(responses_aggregated, on=["Year", "Semester"], how="right")

print(new_merged[['Year', 'Semester', 'Responses', 'Enrollment']])
```
